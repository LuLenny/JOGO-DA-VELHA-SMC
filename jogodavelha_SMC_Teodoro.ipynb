O Segundo, executa uma simulação de 1000 jogos da velha sem nenhuma alteração o resultado portanto é aleatório, varia entre vitoria para Jogador 1 ou Jogador2 com empates a esmo, a utilização de Minimax está presente no penúltimo e último resultado, onde no último foi refinado para apenas um jogador se aproveitar da técnica e vencer predominantemente e obter pouquissimos empates, no penúltimo com ambos utilizando da técnica, empataram em todos os jogos pois estavam constantemente prevendo um a outro e portanto sempre chegavam em "velha".

```python
import random
import matplotlib.pyplot as plt

def display_board(board):
    """Função para exibir o tabuleiro"""
    print('-------------')
    print('| ' + board[1] + ' | ' + board[2] + ' | ' + board[3] + ' |')
    print('-------------')
    print('| ' + board[4] + ' | ' + board[5] + ' | ' + board[6] + ' |')
    print('-------------')
    print('| ' + board[7] + ' | ' + board[8] + ' | ' + board[9] + ' |')
    print('-------------')

def choose_first():
    """Função para escolher aleatoriamente qual jogador começa"""
    if random.randint(0, 1) == 0:
        return 'Jogador 2'
    else:
        return 'Jogador 1'

def space_check(board, position):
    """Função para verificar se uma posição no tabuleiro está livre"""
    return board[position] == ' '

def full_board_check(board):
    """Função para verificar se o tabuleiro está cheio"""
    for i in range(1, 10):
        if space_check(board, i):
            return False
    return True

def win_check(board, mark):
    """Função para verificar se algum jogador ganhou"""
    return ((board[1] == board[2] == board[3] == mark) or
            (board[4] == board[5] == board[6] == mark) or
            (board[7] == board[8] == board[9] == mark) or
            (board[1] == board[4] == board[7] == mark) or
            (board[2] == board[5] == board[8] == mark) or
            (board[3] == board[6] == board[9] == mark) or
            (board[1] == board[5] == board[9] == mark) or
            (board[3] == board[5] == board[7] == mark))

def minimax(board, depth, is_maximizing, player, opponent):
    """Função Minimax"""
    if win_check(board, player):
        return 10 - depth
    if win_check(board, opponent):
        return depth - 10
    if full_board_check(board):
        return 0

    if is_maximizing:
        best_score = -float('inf')
        for i in range(1, 10):
            if space_check(board, i):
                board[i] = player
                score = minimax(board, depth + 1, False, player, opponent)
                board[i] = ' '
                best_score = max(score, best_score)
        return best_score
    else:
        best_score = float('inf')
        for i in range(1, 10):
            if space_check(board, i):
                board[i] = opponent
                score = minimax(board, depth + 1, True, player, opponent)
                board[i] = ' '
                best_score = min(score, best_score)
        return best_score

def best_move(board, player):
    """Função para encontrar o melhor movimento"""
    best_score = -float('inf')
    move = 0
    for i in range(1, 10):
        if space_check(board, i):
            board[i] = player
            score = minimax(board, 0, False, player, 'O' if player == 'X' else 'X')
            board[i] = ' '
            if score > best_score:
                best_score = score
                move = i
    return move

def random_move(board):
    """Função para realizar um movimento aleatório"""
    move = 0
    while move not in range(1, 10) or not space_check(board, move):
        move = random.randint(1, 9)
    return move

def run_game():
    """Função para rodar uma única partida"""
    the_board = [' '] * 10
    turn = choose_first()
    game_on = True

    while game_on:
        if turn == 'Jogador 1':
            position = best_move(the_board, 'X')
            the_board[position] = 'X'

            if win_check(the_board, 'X'):
                return 'Jogador 1'
            elif full_board_check(the_board):
                return 'Empate'
            else:
                turn = 'Jogador 2'
        else:
            position = random_move(the_board)
            the_board[position] = 'O'

            if win_check(the_board, 'O'):
                return 'Jogador 2'
            elif full_board_check(the_board):
                return 'Empate'
            else:
                turn = 'Jogador 1'

def monte_carlo_simulation(num_games):
    """Função principal para rodar a simulação de Monte Carlo"""
    results = {'Jogador 1': 0, 'Jogador 2': 0, 'Empate': 0}
    results_list = []

    for i in range(num_games):
        result = run_game()
        results[result] += 1
        results_list.append(results.copy())

    print(f'Resultados após {num_games} partidas:')
    print(f"Jogador 1 venceu: {results['Jogador 1']} vezes")
    print(f"Jogador 2 venceu: {results['Jogador 2']} vezes")
    print(f"Empates: {results['Empate']} vezes")

    # Plotando os resultados
    plot_results(results_list, num_games)

def plot_results(results_list, num_games):
    """Função para plotar os resultados da simulação"""
    games = list(range(1, len(results_list) + 1))
    jogador1_wins = [result['Jogador 1'] for result in results_list]
    jogador2_wins = [result['Jogador 2'] for result in results_list]
    empates = [result['Empate'] for result in results_list]

    # Proporção acumulada
    jogador1_ratio = [win / i for i, win in enumerate(jogador1_wins, 1)]
    jogador2_ratio = [win / i for i, win in enumerate(jogador2_wins, 1)]
    empate_ratio = [draw / i for i, draw in enumerate(empates, 1)]

    plt.figure(figsize=(14, 7))

    # Gráfico de Proporções Acumuladas
    plt.subplot(1, 2, 1)
    plt.plot(games, jogador1_ratio, label='Jogador 1 (Minimax)', color='blue')
    plt.plot(games, jogador2_ratio, label='Jogador 2 (Aleatório)', color='red')
    plt.plot(games, empate_ratio, label='Empates', color='green')
    plt.xlabel('Número de Jogos')
    plt.ylabel('Proporção Acumulada')
    plt.title('Proporção Acumulada de Vitórias e Empates')
    plt.legend()
    plt.grid(True)

    # Gráfico de Diferença de Vitórias
    jogador1_diff = [jogador1_wins[i] - jogador2_wins[i] for i in range(num_games)]
    plt.subplot(1, 2, 2)
    plt.plot(games, jogador1_diff, label='Diferença de Vitórias (Jogador 1 - Jogador 2)', color='purple')
    plt.xlabel('Número de Jogos')
    plt.ylabel('Diferença de Vitórias')
    plt.title('Diferença de Vitórias entre Jogador 1 e Jogador 2')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Rodar a Simulação de Monte Carlo com 1000 partidas
monte_carlo_simulation(1000)
```
